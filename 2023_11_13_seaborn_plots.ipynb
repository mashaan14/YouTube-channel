```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import datetime
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
g = sns.violinplot(data=df, x="class", y="age")
```

```python
g = sns.violinplot(data=df, x="class", y="age")
date_string = datetime.datetime.now().strftime("%Y-%m-%d--%H-%M-%S-%f")
g.figure.savefig('plot-'+date_string+'.png', bbox_inches='tight', dpi=600)
```

```python
g = sns.violinplot(data=df, x="class", y="age")
g.tick_params(axis='both',which='both',bottom=False,top=False,left=False,right=False,
                labelbottom=False,labeltop=False,labelleft=False,labelright=False);
g.set(xlabel=None, ylabel=None)
date_string = datetime.datetime.now().strftime("%Y-%m-%d--%H-%M-%S-%f")
g.figure.savefig('plot-'+date_string+'.png', bbox_inches='tight', dpi=600)
```

```python
g = sns.violinplot(data=df, x="class", y="age", hue="who")
date_string = datetime.datetime.now().strftime("%Y-%m-%d--%H-%M-%S-%f")
g.figure.savefig('plot-'+date_string+'.png', bbox_inches='tight', dpi=600)
```

```python
g = sns.catplot(data=df, x="who", y="age", col="class", kind="violin")
date_string = datetime.datetime.now().strftime("%Y-%m-%d--%H-%M-%S-%f")
g.figure.savefig('plot-'+date_string+'.png', bbox_inches='tight', dpi=600)
```
